```python
import plotly
import plotly.graph_objs as go
import numpy as np
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot


init_notebook_mode(connected=True)

a = []
b = []
c = []
text_file = open("scattered_rays.txt", "r")
lines = text_file.readlines()
text_file.close()
for i in range(len(lines)):
    values = lines[i].replace('\n','').split(',')
    a.append(values[0])
    b.append(values[1])
    c.append(values[2])

d = []
e = []
f = []
text_file = open("scattered_state.txt", "r")
lines = text_file.readlines()
text_file.close()
for i in range(len(lines)):
    values = lines[i].replace('\n','').split(',')
    d.append(values[0])
    e.append(values[1])
    f.append(values[2])
    
g = []
h = []
j = []
text_file = open("scattered_sphere.txt", "r")
lines = text_file.readlines()
text_file.close()
for i in range(len(lines)):
    values = lines[i].replace('\n','').split(',')
    g.append(values[0])
    h.append(values[1])
    j.append(values[2])

scattered_rays = go.Scatter3d(
    x=a,
    y=b,
    z=c,
    mode='markers',
    marker=dict(
        size=2,
        color='rgb(34, 0, 255)',                # set color to an array/list of desired values
        #colorscale='Viridis',   # choose a colorscale
        opacity=0.6
    )
)

scattered_state = go.Scatter3d(
    x=d,
    y=e,
    z=f,
    mode='markers',
    marker=dict(
        size=2,
        color='rgb(255, 0, 222)',                # set color to an array/list of desired values
        #colorscale='Viridis',   # choose a colorscale
        opacity=0.8
    )
)

scattered_sphere = go.Scatter3d(
    x=g,
    y=h,
    z=j,
    mode='markers',
    marker=dict(
        size=2,
        color='rgb(255, 0, 0)',                # set color to an array/list of desired values
        #colorscale='Viridis',   # choose a colorscale
        opacity=0.6
    )
)

light1 = go.Mesh3d(       
    x= [32,32,68,68],
    y= [81.6,81.6,81.6,81.6],
    z= [63,96,63,96],
    color='rgb(235, 149, 0)',
    opacity=0.8,
    delaunayaxis = 'y'
)

light2 = go.Mesh3d(       
    x= [88,88,99,99],
    y= [0.01,0.01,0.01,0.01],
    z= [63,88,63,88],
    color='rgb(235, 149, 0)',
    opacity=0.8,
    delaunayaxis = 'y'
)

left_green = go.Mesh3d(       
    x= [1,1,1,1],
    y= [0,0,81.6,81.6],
    z= [0,170,0,170],
    color='rgb(0, 255, 0)',
    opacity=0.3,
    delaunayaxis = 'x'
)

right_red = go.Mesh3d( 
    x= [99,99,99,99],
    y= [0,0,81.6,81.6],
    z= [0,170,0,170],
    color='rgb(255, 0, 0)',
    opacity=0.3,
    delaunayaxis = 'x'
)

line1 = go.Scatter3d(
    x=[1, 99],
    y=[81.6, 81.6],
    z=[0, 0],
    marker=dict(
        size=1,
        color='rgb(1,1,1)'
    )
)

line2 = go.Scatter3d(
    x=[1, 99],
    y=[81.6, 81.6],
    z=[170, 170],
    marker=dict(
        size=1,
        color='rgb(1,1,1)'
    )
)

#data = [left_green, right_red, light, line1, line2,scattered_rays,scattered_sphere]
data = [light2,scattered_rays, scattered_state, line1, line2, left_green, right_red]

layout = go.Layout(
    margin=dict(
        l=0,
        r=0,
        b=0,
        t=0
    )
)
fig = go.Figure(data=data, layout=layout)
plotly.offline.plot(fig, filename='3d-scatter-colorscale')
```
